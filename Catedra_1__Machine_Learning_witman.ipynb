```python
import os, requests, zipfile, pandas as pd
from IPython.display import Audio, display

# === Par치metros ===
URL = 'https://media.githubusercontent.com/media/witman92/Catedra-1-_Machine_Learning-no-supervisado/refs/heads/main/5.-Dataset%20audio%20con%204%20horas%20de%20espa%C3%B1ol%20americano%205159%20enunciados%20y%2010%20hablantes.?download=true'
ZIP, DEST = 'dataset_audio.zip', './descargas/audio_extraido'
WAV_DIR, EXT, MAX = os.path.join(DEST, 'WAV'), '.wav', 10

def descargar(url, path):
    if os.path.exists(path): return print(f"游늬 '{path}' ya existe. Descarga omitida.") or True
    print("拘勇 Descargando...")
    try:
        r = requests.get(url, stream=True); r.raise_for_status()
        with open(path, 'wb') as f: [f.write(c) for c in r.iter_content(8192)]
        print(f"九 Descargado: {path}"); return True
    except Exception as e: print(f"仇 Error al descargar: {e}"); return False

def extraer(zip_path, destino):
    if os.path.exists(destino): return print("游늭 Carpeta ya existe. Extracci칩n omitida.") or True
    print("游닍 Extrayendo...")
    try:
        with zipfile.ZipFile(zip_path, 'r') as z: z.extractall(destino)
        print(f"九 Extra칤do en: {destino}"); return True
    except Exception as e: print(f"仇 Error al extraer: {e}"); return False

def crear_df_audios(ruta, ext=EXT, max_audios=MAX):
    data = {
        sub: sorted([f for f in os.listdir(os.path.join(ruta, sub)) if f.endswith(ext)])[:max_audios]
        for sub in sorted(os.listdir(ruta)) if os.path.isdir(os.path.join(ruta, sub))
    }
    if not data: return pd.DataFrame()
    max_len = max(map(len, data.values()))
    return pd.DataFrame({k: v + [None]*(max_len - len(v)) for k, v in data.items()})

def reproducir(df, base, limite=5):
    if df.empty: return print("丘멆잺 No hay audios para reproducir.")
    print(f"\n游댉 Reproduciendo los primeros {limite} audios:\n")
    count = 0
    for sub in df.columns:
        for f in df[sub].dropna():
            if count >= limite: return
            ruta = os.path.join(base, sub, f)
            if os.path.isfile(ruta):
                print(f"游늬 {sub} | 游꿏勇 {f}")
                display(Audio(filename=ruta))
                count += 1

def main():
    if descargar(URL, ZIP) and extraer(ZIP, DEST):
        print(f"游늭 Archivos cargados desde: {WAV_DIR}")
        df = crear_df_audios(WAV_DIR)
        if not df.empty:
            print("\n游꿚 Tabla de audios por subcarpeta:\n"); display(df)
            reproducir(df, WAV_DIR)
        else: print("丘멆잺 No se encontraron archivos .wav.")

if __name__ == "__main__":
    main()
```

## 10.- Dataset-Clasificaci칩n de im치genes de Intel 150x150 en 6 categor칤as.edificios 0,bosque 1, glaciar 2, monta침a 3, mar 4, calle 5

```python
import os, zipfile, requests, pandas as pd, base64
from PIL import Image
from io import BytesIO
from IPython.display import display, HTML

URL_ZIP = ("https://media.githubusercontent.com/media/witman92/"
           "Catedra-1-_Machine_Learning-no-supervisado/refs/heads/main/"
           "4.-%20Dataset-Clasificaci%C3%B3n%20de%20im%C3%A1genes%20de%20Intel?download=true")
ZIP_PATH, EXTRACT_DIR = "repositorio.zip", "dataset"
BASE_PATH = os.path.join(EXTRACT_DIR, "seg_train", "seg_train")

def obtener_info_imagen(ruta):
    try:
        with Image.open(ruta) as img:                              # Abre la imagen
            buffer = BytesIO()
            img.save(buffer, format="JPEG")                        # Guarda imagen en buffer JPEG
            img_b64 = base64.b64encode(buffer.getvalue()).decode()# Codifica imagen en base64
            return img.size + (f"data:image/jpeg;base64,{img_b64}",)
    except:
        return None, None, None                                    # Retorna None si error

def descargar_extraer(url, zip_path, dest_dir):
    try:
        with requests.get(url, stream=True) as r:
            r.raise_for_status()
            with open(zip_path, 'wb') as f:
                for chunk in r.iter_content(8192): f.write(chunk)  # Descarga en chunks
        with zipfile.ZipFile(zip_path) as z:
            z.extractall(dest_dir)                                # Extrae archivo ZIP
        print("九 Descarga y extracci칩n completadas.")
        return True
    except Exception as e:
        print(f"仇 Error: {e}")
        return False

def crear_dataframe(base_path, max_imgs=5):
    if not os.path.exists(base_path):
        print(f"丘멆잺 Ruta no existe: {base_path}")
        return pd.DataFrame()
    datos = {
        cat: [obtener_info_imagen(os.path.join(base_path, cat, f))           # Obtiene info im치genes
              for f in sorted(os.listdir(os.path.join(base_path, cat)))
              if f.lower().endswith(('.jpg','.jpeg','.png'))][:max_imgs]
        for cat in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, cat))
    }
    if not datos:
        print("丘멆잺 No se encontraron im치genes.")
    else:
        print("九 DataFrame generado.")
    return pd.DataFrame({k: pd.Series(v) for k, v in datos.items()})      # Crea DataFrame

def mostrar_imagenes(df):
    if df.empty:
        print("No hay im치genes para mostrar.")
        return

    def a_html(info):
        if not info or len(info) != 3 or info[2] is None:
            return ''
        w, h, b64 = info
        return (f'<div style="text-align:center;"><img src="{b64}" width="100">'
                f'<br><span style="font-size:10px;color:#555;">{w}x{h}px</span></div>')  # Formatea HTML

    df_html = df.apply(lambda col: col.apply(a_html))  # Aplica funci칩n por columna para evitar warning
    display(HTML(df_html.to_html(escape=False)))       # Muestra tabla HTML con im치genes

def main():
    if descargar_extraer(URL_ZIP, ZIP_PATH, EXTRACT_DIR):
        df = crear_dataframe(BASE_PATH)
        mostrar_imagenes(df)
    else:
        print("游띔 No se pudo completar la descarga y extracci칩n.")

if __name__ == "__main__":
    main()
```

Resumen t칠cnico del dataset "Clasificaci칩n de im치genes de Intel":

Este dataset contiene im치genes a칠reas clasificadas en 6 categor칤as (edificios, bosque, glaciar, monta침a, mar y calle), con resoluci칩n de 150x150 p칤xeles. Aunque etiquetado, es 칰til para tareas de aprendizaje no supervisado en visi칩n por computadora.

### Vistos en clase

1. **K-Means Clustering**  
   - Agrupa datos en k cl칰steres minimizando la suma de distancias cuadradas a los centroides.  
   - Algoritmo iterativo: asigna puntos al centroide m치s cercano y recalcula centroides hasta convergencia.  
   - Aplicaci칩n: sobre vectores de texto (ej. TF-IDF) para agrupar causas y acciones similares.

2. **PCA (An치lisis de Componentes Principales)**  
   - Reduce dimensionalidad buscando combinaciones lineales que capturan mayor varianza.  
   - Se usa para visualizaci칩n y reducci칩n del espacio vectorial antes del clustering.  
   - Proceso: estandarizar datos, calcular matriz de covarianza, autovalores/vectores, proyectar datos.

### No vistos en clase

3. **Clustering Jer치rquico**  
   - Construye una jerarqu칤a de cl칰steres mediante fusi칩n iterativa (agglomerativo) o divisi칩n (divisivo).  
   - Visualizaci칩n en dendrograma para explorar granularidad de agrupaciones.  
   - 칔til para entender estructura de problemas y soluciones.

4. **DBSCAN**  
   - Agrupa puntos en regiones de alta densidad, detectando tambi칠n puntos ruido (outliers).  
   - Par치metros: 풧 (radio vecindario), minPts (m칤n. puntos en vecindario para ser cl칰ster).
